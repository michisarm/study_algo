# 문제 설명
NxN 크기의 시험관이 있다. 시험관은 1x1 크기의 칸으로 나누어지며, 특정한 위치에는 바이러스가 존재할 수 있다. 모든 바이러스는 1번부터 K번까지의 바이러스 종류 중 하나에 속한다. <br><br>

시험관에 존재하는 모든 바이러스는 1초마다 상, 하, 좌, 우의 방향으로 증식해 나간다. 단, 매 초마다 번호가 낮은 종류의 바이러스부터 먼저 증식한다. 또한 증식 과정에서 특정한 칸에 이미 어떠한 바이러스가 존재한다면, 그 곳에는 다른 바이러스가 들어갈 수 없다. <br><br>

시험관의 크기와 바이러스의 위치 정보가 주어졌을 때, S초가 지난 후에 (X,Y)에 존재하는 바이러스의 종류를 출력하는 프로그램을 작성하시오. 만약 S초가 지난 후에 해당 위치에 바이러스가 존재하지 않는다면, 0을 출력한다. 이 때 X와 Y는 각각 행과 열의 위치를 의미하며, 시험관의 가장 왼쪽 위에 해당하는 곳은 (1,1)에 해당한다. <br><br>

예를 들어 다음과 같이 3x3 크기의 시험관이 있다고 하자. 서로 다른 1번, 2번, 3번 바이러스가 각각 (1,1), (1,3), (3,1)에 위치해 있다. 이 때 2초가 지난 뒤에 (3,2)에 존재하는 바이러스의 종류를 계산해보자. <br><br>
[1][0][2] <br>
[0][0][0] <br>
[3][0][0] <br><br>


1초가 지난 후에 시험관의 상태는 다음과 같다. <br>
[1][1][2] <br>
[1][0][2] <br>
[3][3][0] <br><br>


2초가 지난 후에 시험관의 상태는 다음과 같다. <br>
[1][1][2] <br>
[1][1][2] <br>
[3][3][2] <br><br>


결과적으로 2초가 지난 뒤에 (3,2)에 존재하는 바이러스의 종류는 3번 바이러스다. 따라서 3을 출력하면 정답이다.


# 입출력 예
첫째 줄에 자연수 N, K가 공백을 기준으로 구분되어 주어진다. (1 ≤ N ≤ 200, 1 ≤ K ≤ 1,000) 둘째 줄부터 N개의 줄에 걸쳐서 시험관의 정보가 주어진다. <br>
각 행은 N개의 원소로 구성되며, 해당 위치에 존재하는 바이러스의 번호가 공백을 기준으로 구분되어 주어진다. <br>
단, 해당 위치에 바이러스가 존재하지 않는 경우 0이 주어진다. 또한 모든 바이러스의 번호는 K이하의 자연수로만 주어진다. <br>
N+2번째 줄에는 S, X, Y가 공백을 기준으로 구분되어 주어진다. (0 ≤ S ≤ 10,000, 1 ≤ X, Y ≤ N) <br>

- 예제 입력 1 <br>
3 3 <br>
1 0 2 <br>
0 0 0 <br>
3 0 0 <br>
2 3 2 <br><br>

- 예제 출력 1 <br>
3 <br><br>

- 예제 입력 2 <br>
3 3 <br>
1 0 2 <br>
0 0 0 <br>
3 0 0 <br>
1 2 2 <br><br>

- 예제 출력 2 <br>
0 <br><br>



- 입력 해석 <br>
N: 시험관의 크기 <br>
K: 바이러스의 종류 <br>
3:N 3:K <br><br>
초기상태 <br>
[1] [0] [2] <br>
[0] [0] [0] <br>
[3] [0] [0] <br><br>
2초 뒤에 (3,2)에 존재하는 바이러스 종류 출력<br>
2:S 3:X 2:Y <br><br>


출제자의 의도 분석 <br>
1. 주요 학습 목표
- BFS/DFS 활용: 바이러스의 확산을 시뮬레이션하는 과정에서 BFS(Breadth-First Search)나 DFS(Depth-First Search) 알고리즘의 구현 능력을 평가
- 시뮬레이션 구현: 매 초마다 발생하는 바이러스의 확산 과정을 정확히 구현하는 능력
- 우선순위 큐 활용: 바이러스 번호가 낮은 순서대로 처리하는 우선순위 큐의 활용 능력
2. 핵심 평가 요소
- 우선순위에 따른 처리: 번호가 작은 바이러스부터 먼저 증식시켜야 하는 로직 구현
- 동시성 제어: 여러 바이러스가 동시에 증식할 때의 처리 방식
- 경계 조건 처리: 시험관의 경계를 벗어나지 않도록 하는 조건 처리
- 시간 복잡도 관리: 효율적인 알고리즘 설계를 통한 시간 제한 내 문제 해결
3. 문제의 난이도
- 중상급: BFS/DFS의 기본적인 구현뿐만 아니라 우선순위 큐를 활용한 추가적인 로직이 필요
- 실제 코딩 테스트에서 2~3레벨에 해당하는 문제
4. 주의해야 할 점
- 바이러스가 동시에 증식할 때 번호가 작은 바이러스부터 처리
- 이미 바이러스가 있는 칸에는 다른 바이러스가 증식할 수 없음
- S초가 지난 후의 상태를 정확히 계산해야 함
- 이 문제는 BFS/DFS 알고리즘을 활용한 시뮬레이션 문제로, 우선순위 큐를 활용한 효율적인 구현이 요구

```python
from collections import deque

# 입력값 예시
# 3 3
# 1 0 2
# 0 0 0
# 3 0 0
# 2 3 2

# - 입력 해석
# N: 시험관의 크기
# K: 바이러스의 종류
# 3:N 3:K
# 초기상태
# [1] [0] [2]
# [0] [0] [0]
# [3] [0] [0]
# 2초 뒤에 (3,2)에 존재하는 바이러스 종류 출력
# 2:S 3:X 2:Y

# 제출용
# N, K = map(int, input().split())
# # grid = [[0] * (N + 1) for _ in range(N + 1)]
# virus_positions = []
# for row in range(1, N + 1):
#     input_row = [0] + list(map(int, input().split()))
#     print("row", input_row)
#     # 2. 그리드에 행 할당
#     # print("grid", grid)
#     # grid[i] = input_row
#     # 3. 바이러스 위치 저장
#     for col in range(1, N + 1):
#         if input_row[col] != 0:
#             virus_positions.append((input_row[col], row, col))
# S, X, Y = map(int, input().split())

# 테스트용
input_value = """
3 3
1 0 2
0 0 0
3 0 0
2 3 2
"""
lines = input_value.strip().split("\n")
# 첫 번째 줄: N, K
N, K = map(int, lines[0].split())
grid = [[0] * (N + 1) for _ in range(N + 1)]
# 바이러스 위치 저장 리스트
virus_positions = []
# 2~N+1줄: 그리드 정보
for row in range(1, N + 1):
    input_row = [0] + list(map(int, lines[row].split()))
    grid[row] = input_row
    # 바이러스 위치 저장
    for col in range(1, N + 1):
        if input_row[col] != 0:
            virus_positions.append((input_row[col], row, col))
# 마지막 줄: S, X, Y
S, X, Y = map(int, lines[-1].split())

# 그리드 출력
print("grid", grid)
print("virus_positions", virus_positions)


def solution():
    #     queues = [
    #       deque(),        # 0번 인덱스
    #       deque(),        # 1번 바이러스 큐
    #       deque(),        # 2번 바이러스 큐
    #       deque()         # 3번 바이러스 큐
    #     ]
    queues = [deque() for _ in range(K + 1)]

    # virus_number : 바이러스 번호, x, y : 위치
    for virus_number, x, y in virus_positions:
        # (second, x, y)
        queues[virus_number].append((0, x, y))
        # virus_positions : [(1, 1, 1), (2, 1, 3), (3, 3, 1)]
        # queues[1].append((0, 1, 1))  # 1번 바이러스 큐에 (0,1,1) 추가
        # queues[2].append((0, 1, 3))  # 2번 바이러스 큐에 (0,1,3) 추가
        # queues[3].append((0, 3, 1))  # 3번 바이러스 큐에 (0,3,1) 추가

    dx = [-1, 1, 0, 0]
    dy = [0, 0, -1, 1]

    for second in range(0, S):
        print("Second: ", second, ", grid: ", grid)
        for virus_number in range(1, K + 1):
            # Process all cur₩rent positions for this virus
            for _ in range(len(queues[virus_number])):
                second, x, y = queues[virus_number].popleft()

                for i in range(4):
                    nx, ny = x + dx[i], y + dy[i]
                    # 1 <= nx <= N : 새 x 좌표가 그리드의 왼쪽 경계를 벗어나지 않는지 확인
                    # 1 <= ny <= N : 새 y 좌표가 그리드의 위쪽 경계를 벗어나지 않는지 확인
                    # grid[nx][ny] == 0: 해당 위치가 아직 바이러스에 감염되지 않았는지 확인
                    if 1 <= nx <= N and 1 <= ny <= N and grid[nx][ny] == 0:
                        grid[nx][ny] = virus_number
                        # 감염시킴
                        queues[virus_number].append((second + 1, nx, ny))
                        print(
                            "virus_number:",
                            virus_number,
                            ", second:",
                            second,
                            ", queues[" + str(virus_number) + "]:",
                            queues[virus_number],
                        )

    print("Second: ", second + 1, ", grid: ", grid)
    print(
        "출력: grid[" + str(X) + "][" + str(Y) + "]: ",
        grid[X][Y] if grid[X][Y] != 0 else 0,
    )


solution()
```
